### 3. Plot Distributions

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE estimates the **Probability Density Function** for one or more variables. 

```python
sns.kdeplot(data=grades, x="class1",shade=True)
```

A **histogram** is another method for visualizing the distribution of data. Bins are created (usually of same width) and then a count is made of the number of points falling within the bin. Bins can be adjusted.

```python
sns.histplot(data=grades, x="class1",bins=15)
```

Below we see a plot of both histogram and kde. 

```python
sns.displot(data=grades,x="class1",kde=True)
```

### 4. Compare Distributions

```python
fig, axes = plt.subplots(1,1, figsize=(12,6))
sns.kdeplot(data=grades, x='class1',fill=True, label="Class1")
plt.axvline(grades['class1'].mean(), linestyle='-.')
sns.kdeplot(data=grades, x='class2',fill=True, label="Class2")
plt.axvline(grades['class2'].mean(), linestyle='-.', color='maroon')
sns.kdeplot(data=grades, x='class3',fill=True, label="Class3" )
plt.axvline(grades['class3'].mean(), linestyle='-.', color='green')
plt.xlabel("Grade")
plt.ylim(0,0.045)
plt.legend()
```

### A Note on Probabilities

What does standard deviation tell us? Yes. Spread, but so what?

```python
print(grades['class1'].mean())
print(grades['class1'].std())
```

Chebyshev's inequality states that a minimum of 75% of values must lie within two standard deviations of the mean and 89% within three standard deviations for a broad range of different probability distributions.

```python
def cheby(k):
    pctg = 1 - 1/k**2
    return pctg
```

```python
cheby(1.5)
```
